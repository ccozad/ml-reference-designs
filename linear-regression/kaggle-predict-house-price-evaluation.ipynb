# Replicate the model class

PyTorch models are all code based and there isn't a great way to save and load the model without the model source code. Investigate ONYX models and Torch Script as a workaround for this problem. For now we copy pasta.

```python
from torch import nn

input_size = 24
output_size = 1

class LinearRegression(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(input_size, output_size)

    def forward(self, x):
        output = self.linear(x)
        return output

    def training_step(self, batch):
        inputs, targets = batch
        output = self(inputs)
        loss = nn.functional.l1_loss(
            input = output,
            target = targets,
            size_average = None,
            reduce = None,
            reduction = 'mean'
        )
        return loss

    def validation_step(self, batch):
        inputs, targets = batch
        output = self(inputs)
        loss = nn.functional.l1_loss(
            input = output,
            target = targets,
            size_average = None,
            reduce = None,
            reduction = 'mean'
        )
        return {'val_loss': loss.detach()}

    def validation_epoch_end(self, outputs):
        batch_losses = [x['val_loss'] for x in outputs]
        epoch_loss = torch.stack(batch_losses).mean()
        return {'val_loss': epoch_loss.item()}

    def epoch_end(self, epoch, result, num_epochs):
        if(epoch) % 10 == 0 or epoch == num_epochs - 1:
            print(f"Epoch {epoch}, val_loss: {result['val_loss']:.4f}")
```

```python
model_path = "kaggle_predict_house_prices_model.pth"

import torch

device = "cpu"
if torch.cuda.is_available():
    device = "cuda"
elif torch.backends.mps.is_available():
    device = "mps"

print(f"Using device {device}")

model = LinearRegression().to(device)
model.load_state_dict(torch.load(model_path))
model.eval()
```

```python
input_columns = training_columns[1:]

inputs_array = test_features_df[input_columns].to_numpy()

sale_prices = []

for values in inputs_array:
    data = torch.Tensor(values).to(device)
    sale_price = model(data);
    sale_prices.append(sale_price.item())

test_features_df["SalePrice"] = sale_prices
test_features_df.head()
```

```python
test_features = evaluation_standard_scaler.inverse_transform(test_features_df.values)

final_features_df = pd.DataFrame(test_features, index=test_features_df.index, columns=test_features_df.columns)
final_features_df["Id"] = final_features_df["Id"].astype(int)

final_features_df[["Id", "SalePrice"]].to_csv("kaggle-final.csv", index=False)
```
