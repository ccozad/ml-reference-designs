# Splitting and merging color channels

```python
image_flowers_bgr = cv.imread("data/flowers.png", cv.IMREAD_COLOR)
b, g, r = cv.split(image_flowers_bgr)

plt.figure(figsize=[20,5])

plt.subplot(141)
plt.imshow(r, cmap="gray")
plt.title("Red Channel")

plt.subplot(142)
plt.imshow(g, cmap="gray")
plt.title("Green Channel")

plt.subplot(143)
plt.imshow(b, cmap="gray")
plt.title("Blue Channel")

image_merged = cv.merge((b, g, r))
plt.subplot(144)
plt.imshow(image_merged[:, :, ::-1])
plt.title("Merged Output")
```

# Change to different color spaces

```python
image_hsv = cv.cvtColor(image_flowers_bgr, cv.COLOR_BGR2HSV)
image_rgb = cv.cvtColor(image_flowers_bgr, cv.COLOR_BGR2RGB)

h, s, v = cv.split(image_hsv)

plt.figure(figsize=[20,5])

plt.subplot(141)
plt.imshow(h, cmap="gray")
plt.title("Hue Channel")

plt.subplot(142)
plt.imshow(s, cmap="gray")
plt.title("Saturation Channel")

plt.subplot(143)
plt.imshow(v, cmap="gray")
plt.title("Value Channel")

plt.subplot(144)
plt.imshow(image_rgb)
plt.title("Original")
```

# Modify individual channel

```python
h_new = h + 15
image_flowers_merged = cv.merge((h_new, s, v))
image_flowers_rgb = cv.cvtColor(image_flowers_merged, cv.COLOR_HSV2RGB)

plt.figure(figsize=[20,5])

plt.subplot(141)
plt.imshow(h_new, cmap="gray")
plt.title("Hue New Channel")

plt.subplot(142)
plt.imshow(s, cmap="gray")
plt.title("Saturation Channel")

plt.subplot(143)
plt.imshow(v, cmap="gray")
plt.title("Value Channel")

plt.subplot(144)
plt.imshow(image_flowers_rgb)
plt.title("Modified")
```

# Save images

```python
cv.imwrite("flowers_SAVED.png", cv.cvtColor(image_flowers_rgb, cv.COLOR_RGB2BGR))
Image(filename="flowers_SAVED.png")
```
